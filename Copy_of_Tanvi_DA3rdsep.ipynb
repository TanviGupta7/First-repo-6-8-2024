```python
import pandas as pd
import numpy as numpy
df=pd.read_csv('/basics_data (1).csv')
df
```

```python
#Mathod2- Imputation
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=numpy.nan,strategy='mean')
imputer.fit(df.iloc[:,1:3])
df.iloc[:,1:3]=imputer.transform(df.iloc[:,1:3])#col F1 include till F3, F3 exclude
df
#
```

```python
#Handle Categorical data
df['Target'].value_counts()
```

```python
#Label Encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Target']=le.fit_transform(df['Target'])
df
```

```python
le.classes_
```

```python
#Ordinal- hqas order,  NOMINAL- doesn't have order
#label encoder used for target while one hot encoder is used for country and color
#one hot encoder
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
df1=pd.DataFrame(ohe.fit_transform(df[['Color']]).toarray(),
columns=['Color_'+str(i) for i in range(len(df['Color'].value_counts()))])
df2=pd.concat([df,df1],axis=1)# initially color were 0,1,2 signifying 3 colors in table now we concat values
df3=df2.drop(['Color'],axis=1)
df3
```

```python
#one hot encoder
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
df1=pd.DataFrame(ohe.fit_transform(df[['Country']]).toarray(),
columns=['Country_'+str(i) for i in range(len(df['Country'].value_counts()))])
df4=pd.concat([df3,df1],axis=1)# initially color were 0,1,2 signifying 3 colors in table now we concat values
df5=df4.drop(['Country'],axis=1)
df5
#label encoding- puts 0,1 value for india, uk but we do not know which one should be 0/1 neither it is target for which we can set the value for.
#we use one hot shot - it transforms country into 2 columns 1, 2 bec 2 countries ind, uk is given. it puts values as- when ind=1 uk =0 and vice versa for the netire value
```

```python
y=df4['Target']
x=df5.drop(['Target'],axis=1)
print(x.shape)
print(y.shape)
```

```python
len(df['Color'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=123)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
#Data frame- x(input), y(output)- xtest, train same for y
#11 features*
```

```python
print(x_test)
```

```python
#Feature Scaling method-1 - Standardization
# x0= (x-mean)/variance
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
print(x_train_sc)
print(x_test_sc)
```

```python
#method-2 normalization
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x_train_mms=mms.fit_transform(x_train)
x_test_mms=mms.transform(x_test)
print(x_train_mms)
print(x_test_mms)
```

```python
#Feature extraction
#Method-1, PCA
from sklearn.decomposition import PCA
#11 features out of which 2 have to be selected
pca=PCA(n_components=2)
x_train_pca=pca.fit_transform(x_train)#unsupervised learning where target(y) is not required
x_test_pca=pca.transform(x_test)
print(x_train_pca)
print(x_test_pca)
print(pca.explained_variance_ratio_)# print lambda
#find eigen value lambda , pca sort the lambda in descending order
#dimension reduction(reducing dimension of lambda) - 1)feature selection- as it is data goes 2) feature extraction(reduces dimension ie reduces perpendicular dimension of lambda )-i)pca, ii)lda
#llm-large lang. transformer

#tablet instead of food- same energy : this is the role of dimension reduction
```

```python
#Method-2, Linear Discriminant Analysis LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA(n_components=1)
x_train_lda=lda.fit_transform(x_train,y_train) #supervised learning- target y required ,min(n_features, n_classes-1).
x_test_lda=lda.transform(x_test)
print(x_train_lda)
print(x_test_lda)
```
